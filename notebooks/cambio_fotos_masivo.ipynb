### Lectura de tokens

Ahora que ya tenemos cargada toda la informaci√≥n vamos a cargar los tokens y datos necesarios para comunicarnos con mercado libre ademas de generar las funciones necesarias para hacer el refresh del token.

```python
aplicaciones = db["aplicaciones"]
aplicacion_co = aplicaciones.find_one({"origen": "CO"})
# Leemos la primera aplicaci√≥n para validar la lectura correcta

print(aplicacion_co.get("client_secret"))
print(aplicacion_co.get("token"))
print(aplicacion_co.get("refresh_token"))
print(aplicacion_co.get("seller_id"))
```

```python
def renovar_token(store):
    url = "https://api.mercadolibre.com/oauth/token"
    payload = {
        "grant_type": "refresh_token",
        "client_id": store["app_id"],
        "client_secret": store["client_secret"],
        "refresh_token": store["refresh_token"]
    }
    try:
        resp = requests.post(url, data=payload)
        resp.raise_for_status()
        data = resp.json()
        store["access_token"] = data["access_token"]
        store["refresh_token"] = data["refresh_token"]
        print(f"[{store['origen']}] Token renovado {data["access_token"]}")
        return True
    except Exception as e:
        print(f"[{store['origen']}] ERROR renovar token: {e}")
        return False
```

```python
origen = "DS"
aplicacion = aplicaciones.find_one({"origen": origen})
renovar_token(aplicacion)
ACCESS_TOKEN = aplicacion.get("token")
```

```python
# -----------------------------
# Funciones auxiliares
# -----------------------------
def descargar_imagen(url):
    """Descarga la imagen desde un link y devuelve el contenido en bytes"""
    try:
        resp = requests.get(url, timeout=20, stream=True)
        if resp.status_code == 200:
            return resp.content
        else:
            print(f"‚ùå Error al descargar {url}: {resp.status_code}")
            return None
    except Exception as e:
        print(f"‚ùå Excepci√≥n descargando {url}: {e}")
        return None

def subir_imagen_ml(imagen_bytes):
    """Sube una imagen a Mercado Libre y devuelve el id"""
    try:
        files = {'file': ('imagen.jpg', imagen_bytes)}
        headers = {"Authorization": f"Bearer {ACCESS_TOKEN}"}
        url = "https://api.mercadolibre.com/pictures/items/upload"
        resp = requests.post(url, headers=headers, files=files, timeout=30)
        if resp.status_code in (200, 201):
            data = resp.json()
            return data.get("id")
        else:
            print(f"‚ùå Error subiendo imagen a ML: {resp.status_code} - {resp.text[:200]}")
            return None
    except Exception as e:
        print(f"‚ùå Excepci√≥n subiendo imagen a ML: {e}")
        return None

def actualizar_item_ml(item_id, picture_ids):
    """Actualiza el item en ML con las im√°genes subidas"""
    try:
        url = f"https://api.mercadolibre.com/items/{item_id}"
        headers = {
            "Authorization": f"Bearer {ACCESS_TOKEN}",
            "Content-Type": "application/json",
            "Accept": "application/json"
        }
        payload = {"pictures": [{"id": pid} for pid in picture_ids]}
        resp = requests.put(url, headers=headers, json=payload, timeout=30)
        if resp.status_code == 200:
            print(f"‚úÖ Item {item_id} actualizado con {len(picture_ids)} im√°genes.")
        else:
            print(f"‚ùå Error actualizando item {item_id}: {resp.status_code} - {resp.text[:200]}")
    except Exception as e:
        print(f"‚ùå Excepci√≥n actualizando item {item_id}: {e}")

def validar_url(url):
    """Valida que la URL exista y sea una imagen"""
    if not url or pd.isna(url):
        return False
    try:
        resp = requests.head(url, timeout=10, allow_redirects=True)
        if resp.status_code in (200, 206) and resp.headers.get("Content-Type", "").startswith("image"):
            return True
        else:
            return False
    except:
        return False
```

```python
from time import sleep

# -----------------------------
# Flujo principal
# -----------------------------
for index, row in tiendas_cardic.iterrows():
    ca = row["CA"].strip()
    print(f"\nüîπ Procesando CA: {ca}")

    # Buscar items en MongoDB que tengan el CA en seller_custom_sku y el origen correcto
    query = {"seller_custom_sku": {"$regex": ca}, "origen": origen}
    items = list(items_col.find(query))
    item_ids = [str(item["_id"]) for item in items]
    print(f"üì¶ Se encontraron {len(item_ids)} items para {ca}")

    # Recopilar todas las im√°genes del row
    picture_columns = [f"IMAGNE{i}" for i in range(1, 11)]
    picture_ids = []

    for col in picture_columns:
        url = row.get(col)
        if url and validar_url(url):
            img_bytes = descargar_imagen(url)
            if img_bytes:
                pic_id = subir_imagen_ml(img_bytes)
                if pic_id:
                    picture_ids.append(pic_id)
        elif url:
            print(f"‚ö†Ô∏è URL inv√°lida o inaccesible: {url}")

    # Asignar las im√°genes a cada item
    for item_id in item_ids:
        if picture_ids:
            actualizar_item_ml(item_id, picture_ids)
            sleep(0.5)
        else:
            print(f"‚ö†Ô∏è No hay im√°genes v√°lidas para asignar al item {item_id}")

print("\nüéâ Proceso finalizado para todos los CA.")
```
